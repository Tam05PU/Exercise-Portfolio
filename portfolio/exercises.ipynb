```python
#question 1
# Import sqrt() from the math module
from math import sqrt 
number = int(input("Enter a number: "))  # Ask the user to enter a number
                                        # input() returns text, so int() converts it to an integer
if number <= 1: #chehcs if the number is greater/equal to 1 as prime number are greater than 1
    print(number, "This is not a prime number")  
elif number > 1:
    for i in range(2, int(sqrt(number)) + 1):  
        if number % i == 0:
            print(number, "This is not a prime number")
            break
    else:
        print(number, "This is a prime number")
```

```python
#question5 

# Define a function called min_max
def min_max(numbers):
    # Find the smallest value in the list using min()
    min_value = min(numbers)
    # Find the largest value in the list using max()
    max_value = max(numbers)
    # Return both the minimum and maximum values
    return min_value, max_value
# Create a list of numbers
num = [6, 8, 3, 4, 7]
# Call the function and store the returned values
min_val, max_val = min_max(num)
# Print the minimum value
print("Min value is:", min_val)
# Print the maximum value
print("Max value is:", max_val)
```

```python
#question 6 
import random

random_number_guess = random.randint(1, 10) #generate the number between 1 and 10 

guess = 0  #start value before user input 

while guess != random_number_guess:
    guess = int(input("Guess a number between 1 and 10: ")) #asks user to guess the number between 1 and 10 

    if guess > random_number_guess: #check if the guess number is bigger than generated one
        print("Too high! Try again.")
    elif guess < random_number_guess:#check if the guess number is smaller than generated one
        print("Too low! Try again.")
    else:
        print("Congratulations! You guessed the correct number.")
```

```python
#question 7
# Store the correct password in a variable
user_password = "exercise7"
# Start an infinite loop so the user keeps getting asked  for the password until they enter the correct one
while True:
    # Ask the user to enter a password
    password = input("Enter the password: ")
    # Check if the entered password matches the correct password
    if password == user_password:
        # If the password is correct, display access granted
        print("Access Granted!")
        # Stop the loop because the correct password was entered
        break
    else:
        # If the password is incorrect, display an error message and the while loop will continue
        print("Incorrect password. Try again.")
```

```python
#guestion 8 
import math

def series(x, N):
    total = 0
    for i in range(1, N + 1):
        total += (x * (5**i)**2) / (i + math.factorial(i + 1))
    return round(total, 2)


x = int(input("Enter x: "))
N = int(input("Enter N: "))

print("Result:", series(x, N))
```

```python
#question 9 
import numpy as np
import matplotlib.pyplot as plt
import PIL  # Python Imaging Library - image editing
import urllib
#store the image link
image_url = "https://thumbs.dreamstime.com/b/ocean-sunset-landscape-bird-high-resolution-image-flying-towards-colorful-romantic-sky-246711534.jpg"
#open the image link 
image_data = urllib.request.urlopen(image_url)
image = PIL.Image.open(image_data)
#convert the image into the numpy array
np_image = np.array(image)


horizontal_mirror = np.fliplr(np_image)

# Vertical mirror (top-bottom flip)
vertical_mirror = np.flipud(np_image)

# Display images
plt.figure(figsize=(10,4))

plt.subplot(1,3,1)
plt.title("Original")
plt.imshow(np_image)
plt.axis("off")

plt.subplot(1,3,2)
plt.title("Horizontal Mirror")
plt.imshow(horizontal_mirror)
plt.axis("off")

plt.subplot(1,3,3)
plt.title("Vertical Mirror")
plt.imshow(vertical_mirror)
plt.axis("off")

plt.show()
```

```python
#question 10    
import numpy as np 
# Temperature data in Celsius
temps = [22, 24, 19, 21, 23, 25, 26, 27, 24, 23, 20, 22, 21, 19, 18, 22, 23,
         24, 26, 28, 30, 27, 25, 26, 24, 22, 21, 20, 19, 23]

# 1) Create a NumPy array from the list
temp_array = np.array(temps)
```

```python
# 2) Find minimum, maximum, and average temperature
print("Min:", temp_array.min())
print("Max:", temp_array.max())
print("Average:", round(temp_array.mean(), 2))
```

```python
# 3) Convert Celsius to Fahrenheit: F = (C * 9/5) + 32
fahrenheit_array = (temp_array * 9/5) + 32 
print("Temperatures in Fahrenheit:", fahrenheit_array)

```

```python
# 4) Find all days where temperature was above 25°C
above_25 = temp_array[temp_array > 25]   # filtering using condition
print("Temperatures above 25°C:", above_25)
```

```python
# 5) Count how many days had temperatures below 20°C
below_20_count = np.sum(temp_array < 20)  # counts True values
print("Number of days below 20°C:", below_20_count)
```

```python
# 6) Sort the temperature data in ascending order
sorted_temps = np.sort(temp_array)
print("Sorted temperatures:", sorted_temps)
```

```python
# 7) Reshape the 1D array into a 5x6 2D array (5 weeks, 6 days each)
weeks = temp_array.reshape(5, 6)
print("Temperature data reshaped into 5x6 array:\n", weeks)
```

```python
#question 12
import seaborn as sns
import matplotlib.pyplot as plt

# Load Iris dataset from Seaborn
df_iris = sns.load_dataset("iris")

# Select numerical columns
numeric_columns = df_iris.select_dtypes(include="number").columns

# Generate a boxplot for each numerical column
for column in numeric_columns:
    plt.figure(figsize=(6, 4))
    plt.boxplot(df_iris[column].dropna())
    plt.title(f"Boxplot of {column}")
    plt.ylabel(column)
    plt.grid(True, axis="y", linestyle="--", alpha=0.7)
    plt.tight_layout()
    plt.show()
```

```python
#question 14
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
# Load the tips dataset from seaborn
df_tips = sns.load_dataset("tips")

inputs, targets = df_tips[['total_bill']], df_tips['tip']

# Split the data into training data and testing data,test_size=0.1 means 10% of the data will be used for testing
inputs_train, inputs_test, targets_train, targets_test = train_test_split(inputs, targets, test_size=0.1)

# Create and train the linear regression model using the training data
model = LinearRegression().fit(inputs_train, targets_train)

# Predict tip for a single bill amount
bill_amount = pd.DataFrame({'total_bill': [30.00]})
predictions = model.predict(bill_amount)

# Print the predicted tip, formatted to 2 decimal places
print(f"Predicted tip for a bill of ${30:.2f}: ${predictions[0]:.2f}")
```
