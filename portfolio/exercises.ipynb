```python
#question 1
#asks a user to enter a number

number = int(input("Enter a number:"))  # input() function is fed to the int() function
if number >1:
    for i in range(2, int(number**0.5) + 1):
        if(number%i)==0:
            print(number,"This is not a prime number")
            break
    else:
        print(number,"This is a prime number")
 
```

```python
#question 2
count = int(input("How many numers in Fibonacci set to be calculated and displayed?"))
a, b = 0, 1
for _ in range(count):
    print(a, end=" ")
    a, b = b, a + b
```

```python
# Top half (5 to 1 stars)
n = 5
for i in range(n, 0, -1):
    print("* " * i)
# Bottom half (2 to 5 stars)
for i in range(2,n + 1):
    print("* " * i)
```
